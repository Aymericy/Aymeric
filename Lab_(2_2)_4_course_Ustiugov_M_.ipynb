#**Предобработка и загрузка библиотек**


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

DF_train = pd.read_csv('Train.csv')
DFt = DF_train.drop(columns = ['Class'],axis = 1) # Удаление столбца Class
DFt
```

```python
#Объединение в один Датафрейм
DF_test = pd.read_csv('Test.csv')
DF_test
df_full = pd.concat([DFt,DF_test],sort = False, axis = 0)
df_full
```

```python
#Информация о массиве
from sklearn.preprocessing import StandardScaler

df_full.info()
df_drop = df_full.dropna() # Удаляем пропущенные значения NaN
df_dropa = df_drop.value_counts()
df_dropa

```

```python
#Посмотрим на корреляцию данных между собой
df_drop.corr()
fig, ax = plt.subplots(figsize=(13, 5))
sns.heatmap(df_drop.corr(), annot=True)
#Вывод: Большинство значений коррелирует отлично с периметром!!
```

```python
df_drop.columns
```

```python
#Выделим признаки, которые нам интересны и нормализуем данные с помощью минмакс
from sklearn.preprocessing import MinMaxScaler

claster_tags = ['Extent','Solidity']
X = df_drop[claster_tags]

dfs = X.copy(deep=True)
dfs[claster_tags] = MinMaxScaler().fit_transform(dfs)
dfs
```

#Метод К-means и метод Локтя

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=4, init='k-means++', max_iter=2000, n_init=10, random_state=42)
km.fit(dfs[claster_tags])

y_pred_km = km.predict(dfs[claster_tags])
df_drop['Кластер'] = km.labels_
```

```python
#Визуализация

x= df_drop['Кластер']
y = df_drop['EquivDiameter']
fig, ax = plt.subplots(figsize=(8,5))
sc = ax.scatter(x, y, c=y_pred_km, cmap='tab20')
ax.grid(True)
plt.xlabel("Кластер",fontsize=14)
plt.ylabel("EquivDiameter",fontsize=15)
ax.legend(*sc.legend_elements(), title='Кластеры')
#K-means выделил 4 кластера достаточно неплохо
```

```python
#метод Локтя

criterion = []

for k in range(1,7):
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(dfs[claster_tags])
    criterion.append(kmeans.inertia_)

plt.plot(range(2,8), criterion)
plt.title('Выбор количества кластеров методом локтя')
plt.xlabel('Количество кластеров')
plt.ylabel('WCSS')
#Метод локтя выделил 4 кластера
```

```python
# PCA
from sklearn.decomposition import PCA

dfs[["PC1", "PC2"]] = PCA(n_components=2).fit_transform(dfs)

plt.rcParams["figure.figsize"] = 4, 3

plt.scatter(dfs["PC1"], dfs["PC2"], color="red", marker=".");
#Вывод - понятного мало чего
```

```python
#Оценка метода К-means силуэт
from sklearn.metrics import adjusted_rand_score, silhouette_score
print(f"Силуэт (k-means) = {silhouette_score(dfs[claster_tags], km.labels_)}")
# Метрика силуэт показывает, что разбиение достаточно среднее
#Вывод - Из теории изветсно что 4 кластера, а метрика силуэтов показывает 0,49, что очень средне.
```

#DBSCAN метод

```python
from sklearn.cluster import DBSCAN

dbs_s = DBSCAN(eps=0.03, min_samples=9)
y_pred_dbs_s = dbs_s.fit_predict(dfs[claster_tags])
labels_dbscan = dbs_s.labels_
dbs_s.get_params()
dfd = dfs.copy(deep = True)
df_drop['Кластер1'] = dbs_s.labels_
df_drop.value_counts('Кластер1')
```

```python
#Визуализация
x= df_drop['Кластер1']
y = df_drop['Solidity']
fig, ax = plt.subplots(figsize=(8,5))
sc = ax.scatter(x, y, c=y_pred_dbs_s, cmap='tab20')
ax.grid(True)
plt.xlabel("Кластер",fontsize=14)
plt.ylabel("Solidity",fontsize=15)
ax.legend(*sc.legend_elements(), title='Кластеры')
#DBSCAN выделить кластеры смог, только лишь подобрав в ручную оптимальные параметры
```

```python
#Оценка силуэтом
print(f"Силуэт (dbscan) = {silhouette_score(dfs[claster_tags], dbs_s.labels_)}")
#Силуэты показывают, что разбиение плохое
```

# Стохастическое T-sne и HDBSCAN

```python
from sklearn.manifold import TSNE

dfs.shape
X_embedded = TSNE(n_components=2, learning_rate='auto', init='random', perplexity=7).fit_transform(dfs[claster_tags])
X_embedded.shape

model_km_embedded = KMeans(4, n_init=10)
model_km_embedded.fit(X_embedded)
y_pred_km_embedded = model_km_embedded.predict(X_embedded)

fig, ax = plt.subplots(figsize=(8,5))
sc = ax.scatter(X_embedded[:,0], X_embedded[:,1], c=y_pred_km_embedded, cmap='tab10')
ax.grid(True)
ax.set_xlabel("Синтетический признак 1",fontsize=14)
ax.set_ylabel("Синтетический признак 2",fontsize=15)
ax.legend(*sc.legend_elements(), title='Кластеры')
# T-sne смог выделить в искусственном пространстве по синтетическим признакам 4 кластера достаточно однозначно
```

```python
from sklearn.cluster import HDBSCAN

hdbscan = HDBSCAN(min_cluster_size=10, min_samples=9, cluster_selection_epsilon=0.03)
hdbscan.fit(dfs[claster_tags])
y_pred_hdb = hdbscan.fit_predict(dfs[claster_tags])
labels_hdb = hdbscan.labels_
df_drop['Кластер3'] = hdbscan.labels_
df_drop.value_counts('Кластер3')

x= df_drop['Кластер3']
y = df_drop['Solidity']
fig, ax = plt.subplots(figsize=(8,5))
sc = ax.scatter(x, y, c=y_pred_hdb, cmap='tab20')
ax.grid(True)
plt.xlabel("Кластер",fontsize=14)
plt.ylabel("Solidity",fontsize=15)
ax.legend(*sc.legend_elements(), title='Кластеры')
print(f"Силуэт (HDBSCAN) = {silhouette_score(dfs[claster_tags], hdbscan.labels_)}")
#Улучшенный ДБСКАН выделил, также 4 кластера, однако силуэт показал, что разбиение далеко от правды
```
